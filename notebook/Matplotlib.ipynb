```python
import matplotlib.pyplot as plt
from numpy.random import randn
import numpy as np

# numpy.random.randn(50)  ———— 是指在标准正态分布中返回50个样本值
# numpy.random.rand(50) ————是指在标准正态分布中返回的随机样本位于[0, 1)之间

# 柱状图
ax1 = plt.subplot(2, 2, 1)     # 'k--'表示绘制黑色虚线图
ax1.hist(randn(100), bins=20, color='black', alpha=0.3)     # bins表示图中有几条柱，color表示柱的颜色，alpha表示透明度，0为完全透明


# 散点图
ax2 = plt.subplot(2, 2, 2)
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))


# 折线图
ax3 = plt.subplot(2, 2, 3)
plt.plot(randn(50).cumsum(), 'k--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(100), bins = 20, color='red', alpha = 0.3)
        
plt.subplots_adjust(wspace=0, hspace=0)    # 此代码为修改间距


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn

# 折线图的画法
fig = plt.figure(figsize=(20,5))
plt.subplot(1, 3, 1)
plt.xlim([0, 35])                      # 设置x轴的刻度范围
plt.plot(randn(30).cumsum(), 'k--')     # 'k--'不会标出各个点

plt.subplot(1, 3, 2)
plt.plot(randn(40).cumsum(), 'ko--')    # 'ko--'会标出各个点

plt.subplot(1, 3, 3)
plt.plot(randn(20).cumsum(), 'k--', drawstyle='steps-post')     # 采用drawstep属性可以修改非真实数据间的填值方式

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn

fig = plt.figure(1)
plt.xlabel('Xlabel', fontsize='12')
plt.ylabel('Ylabel', fontsize='12')
plt.title("Statisitcs")
plt.xticks([0, 250, 500, 750, 1000], ['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='12')

        # 可以采用plt.xticks()方法进行修改刻度，和替换刻度标签，rotation表示旋转角度，fontsize表示字体大小

plt.plot(randn(1000).cumsum(), label='one')
plt.legend(loc='best')     # legend表示图例，loc=’best‘参数表示将图例放在一个最不碍事的位置
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn

# 多线合一的图
fig = plt.figure(figsize=(10, 3))
ax = plt.subplot(1, 1, 1)
ax.plot(randn(50).cumsum(), 'k--', color='red', label='one')
ax.plot(randn(50).cumsum(), 'ko--', color='green', label='two')
ax.plot(randn(50).cumsum(), 'k.', color='blue', label='three')
ax.legend(loc='best')
plt.show()

# plt.savefig('threeInOne.png', bbox_inches = 'tight')     # 保存图片
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import randn

# 采用pandas进行高级绘图(Series)
# 线型图
s = pd.Series(randn(10).cumsum(), index = np.arange(0, 100, 10))
s.plot()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import randn


# 采用pandas进行高级绘图(DataFrame)
# 线型图
s = pd.DataFrame(randn(10, 3).cumsum(0),
                 columns=['A', 'B', 'C'],
                 index = np.arange(0, 100, 10))
s.plot()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 采用Pandas  ————  Series绘制柱状图
fig, axes = plt.subplots(3, 1, figsize=(5, 10))
data = pd.Series(np.random.rand(5), index=['a','b','c','d','e'])

data.plot(kind='kde', ax=axes[0], alpha = 0.5)     # kind = 'kde'表示生成密度图
data.plot(kind='bar', ax=axes[1], alpha = 0.5)     # bar 正直方图
data.plot(kind='barh', ax=axes[2], alpha = 0.7)    # barh水平直方图

plt.show()
```
